linear, bar, scatter, boxplot, pie, radar, multiple series chart and 3D charts.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diamonds = pd.read_csv(r'C:\Users\rana_\IronhackDA\DAFT_212\module_2\Project_Week_4\diamonds_clean.csv')
diamonds = diamonds.drop('Unnamed: 0', axis=1)
```

```python
diamonds.describe()
```

histogram plot

```python
ax = diamonds.hist(column='price', bins=25, grid=False, figsize=(10,9),color='#86bf91', zorder=2, rwidth=0.9) 
```

```python
plt.style.use('ggplot')
ax = diamonds.hist(column='price', bins=10, grid=False, figsize=(10,9),color='#86bf91', zorder=2, rwidth=0.9) 
plt.title('Prices of the diamonds')
plt.xlabel('price')
plt.ylabel('count')
plt.show()
```

BAR PLOT

```python
diamonds['cut'].value_counts().plot(kind='bar',figsize=(10,9), color ='#86bf91')
plt.title('Quality of the cut of diamonds')
plt.xlabel('Quality of the cut')
plt.ylabel('count')
plt.show()
```

SCATTER

```python
diamonds.plot.scatter(x='carat', y='price', figsize=(10,9), color ='#86bf91') 
plt.title('Correlation between the carat and the price of the diamond')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```

linear

```python
plt.plot(diamonds["z"], diamonds["price"], color ='#86bf91') 
plt.title('Correlation between the depth and the price of the diamond')
plt.xlabel('Depth')
plt.ylabel('Price')
plt.show() 
```

```python
c =plt.figure(figsize=(10, 9))
plt.title('Correlation between the depth and the price of the diamond')
plt.xlabel('Depth')
plt.ylabel('Price')
sns.regplot(diamonds.z, diamonds.price, scatter=False)

plt.ylim(0,20000)
```

Boxplot

```python
diamonds.boxplot(column="carat",figsize=(10,9)) 
plt.title('Carat boxplot')
#75 % of the diamonds's weight is less thant 1.04 carat
```

```python
diamonds['carat'].median()
```

Pie

```python
diamonds.cut.value_counts().plot(kind='pie', autopct='%1.1f%%',startangle=90, legend = False, figsize=(10,9))
```

Radar

```python
angles=np.linspace(0,2*np.pi,len(diamonds), endpoint=False)
print(angles)
```

```python
fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(polar=True)
ax.plot(angles,diamonds['price'])
plt.show()
```

```python
fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(polar=True)
ax.plot(angles,diamonds['carat'])
plt.show()
```

```python
fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(polar=True)
ax.plot(angles,diamonds['z'])
plt.show()
```

Multiple series-STACKED PLOT

```python
data1=diamonds[['cut','color','clarity_cat', 'carat']].groupby(['cut']).agg("mean")
data1
```

```python
data1.round(1).plot(kind='bar', stacked=True, figsize=(10,8))

plt.xlabel('cut')
plt.ylabel('mean')

plt.title("Carat, clarity, color by cut")
plt.legend()
plt.show()
```

3D chart

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.scatter3D(diamonds['x'], diamonds['y'], diamonds['z']);
ax.set_xlabel("length, mm", fontsize=15)
ax.set_ylabel("width, mm", fontsize=15)
ax.set_zlabel("depth, mm", fontsize=15)
ax.set_title("Diamond dimensions", fontsize=20)
```
